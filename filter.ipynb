```python
file_enron = 'enron-v1.csv'
enron_data = pd.read_csv(file_enron)
jah = enron_data['toEmail'].unique()
jah2 = enron_data['toJobtitle'].unique()
jahh = tuple(jah) 
jahh2 = tuple(jah2)
enron_data
```

```python
ToEmail = dcc.Dropdown(
    id = 'ToEmaildropdown',
    options=jah,
    multi=True,
    value="MTL",
    placeholder = "Select an email adress"
)  

FromEmail = dcc.Dropdown(
    id = 'FromEmaildropdown',
    options=jah,
    multi=True,
    value="MTL",
    placeholder = "Select an email adress"
)  

ToJob = dcc.Dropdown(
    id = 'ToJobdropdown',
    options=jah2,
    multi=True,
    value="MTL",
    placeholder = "Select a job"
)  

FromJob = dcc.Dropdown(
    id = 'FromJobdropdown',
    options=jah2,
    multi=True,
    value="MTL",
    placeholder = "Select a job"
)  

DatePicker = dcc.DatePickerRange(
    id='date-picker-range',
    start_date=date(1998, 11, 12),
    end_date=date(2002, 6, 20),
    end_date_placeholder_text='Select a date'
)

UpdateButton = html.Button('Update', id='button')
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
app.layout = html.Div([
    ToEmail,
    FromEmail,
    ToJob,
    FromJob,
    DatePicker,
    UpdateButton,
])


@app.callback(
    Input('ToEmaildropdown', 'ToEmail'),
    Input('FromEmaildropdown', 'FromEmail'),
    Input('ToJobdropdown', 'ToJob'),
    Input('FromJobdropdown', 'FromJob'),
    Input('date-picker-range', 'Date'),
    Input('button', 'value'))
def update_df(ToEmail, FromEmail, ToJob, FromJob, Date):
    FromEmailF = (enron_data['fromEmail'] == FromEmail)
    ToEmailF = (enron_data['toEmail'] == ToEmail)
    FromJobF = (enron_data['fromJob'] == FromJob)
    ToJobF = (enron_data['toJob'] == ToJob)
    DateF = (enron_data['date'] == Date)
    Filter = (FromEmailF & ToEmailF & FromJobF & ToJobF & DateF)
    enron_data_filtered = enron_data[Filter]
    return enron_data_filtered


if __name__ == '__main__':
    app.run_server(debug=True)
```
